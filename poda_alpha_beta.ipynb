```python
import math

def definicion_estados():
    global sucesores
    global heuristico
    sucesores = {'E1':[{'E11':math.inf,'E12':math.inf,'E13':math.inf}],
                'E11':[{'E111':math.inf,'E112':math.inf}],
                'E12':[{'E113':math.inf,'E114':math.inf}],
                'E13':[{'E115':math.inf}],
                'E111':[{'E1111':math.inf,'E1112':math.inf}],
                'E112':[{'E1113':math.inf}],
                'E113':[{'E1114':math.inf,'E1115':math.inf}],
                'E114':[{'E1116':math.inf}],
                'E115':[{'E1117':math.inf}],
                'E1111':[{'E11111':5,'E11112':6}],
                'E1112':[{'E11113':7,'E11114':4,'E1111_13':5}],
                'E1113':[{'E11115':3}],
                'E1114':[{'E11116':6}],
                'E1115':[{'E11117':6,'E1111_14':9}],
                'E1116':[{'E11118':7}],
                'E1117':[{'E11119':5}],
                'E1118':[{'E1111_10':9,'E1111_11':8}],
                'E1118':[{'E1111_12':6}]
                }
    heuristico = {
                'E11111':[5],
                'E11112':[6],
                'E11113':[7],
                'E11114':[4],
                'E11115':[3],
                'E11116':[6],
                'E11117':[6],
                'E11118':[7],
                'E11119':[5],
                'E1111_10':[9],
                'E1111_11':[8],
                'E1111_12':[6],
                'E1111_13':[5],
                'E1111_14':[9]}
    
    
def initial():
    return 'E1'
    
def actionResults(estado):
    estado = estado.upper()
    try:
        secuencia_sucesores = sucesores[estado]
    except KeyError:
        print(estado,"no es un estado valido")
    return secuencia_sucesores
    
def isTerminal(TState):
    return False

def player(TState):
    player = "MAX" if(len(TState)%2 == 0) else "MIN"
    return player
    
def rival(TPlayer):
    TPlayer.upper()
    player = "MAX" if(TPlayer == "MAX") else "MIN"
    return player

def heuristicEval(TState):
    try:
        estado = TState.upper()
        return heuristico[TState]
    except ValueError:
        print(estado, "no es un estado valido")
        return 0




def minValue(TState, number_d, number_alpha, number_beta):
    if (number_d == 0 or isTerminal(TState)): return heuristicEval(TState)[0]
    value = math.inf

    for state in actionResults(TState)[0]:
        value = min(value, maxValue(state, number_d-1, number_alpha, number_beta))
        number_beta= min (value, number_beta)

        dictionary =sucesores[TState][0]
        dictionary[state]= value
      
        if (number_beta <= number_alpha):      break
    return value

def maxValue(TState, number_d, number_alpha, number_beta):
    if (number_d == 0 or isTerminal(TState)): return heuristicEval(TState)[0]
    value = -math.inf

    for state in actionResults(TState)[0]:
        value = max(value, minValue(state, number_d-1, number_alpha, number_beta))
        number_alpha = max (value, number_alpha)

        dictionary =sucesores[TState][0]
        dictionary[state]= value
      
        if (number_alpha >= number_beta): break
    return value


def alphaBetaDecision(TState, number_d):
    alpha = -math.inf
    beta = math.inf
    value = maxValue(TState, number_d, alpha, beta)
    
    for state in actionResults(TState):
        for llave in state:
            if (state[llave] == value): 
                return llave
    return "no hay camino"


def main():
    definicion_estados()
    print(alphaBetaDecision('E1',4))
main()
```
